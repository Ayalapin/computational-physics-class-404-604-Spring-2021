```python
import numpy as np

def derifw(y,cosx,dx):
    return (y(x+dx)-y(x))/dx


import numpy as np


print('forward difference',derifw(np.sin,x,dx))
```

```python
import numpy as np
import math 

x=math.pi/4
dxarr=[]
yarr=[]
for dx in 10**np.arange(-16, -0.5, 0.1):
    y=derifw(np.sin,x,dx)
    print('dx %g,  y %g'%(dx,y))
    dxarr.append(dx)
    yarr.append(y)
```

```python
import numpy as np
import matplotlib.pyplot as plt

plt.plot(dxarr, abs(yarr-np.cos(1.)), 'b-', lw=2)     
plt.xscale('log')
plt.yscale('log')
plt.xlabel('dx')
plt.ylabel('Error')
plt.show()
```

```python
0.1 is the optimal value 
```

```python
import numpy as np

def dericd(y,cosx,dx):
    return (y(x+dx/2)-y(x - dx/2))/dx


import numpy as np


print('central difference',dericd(np.sin,x,dx))
```

```python
import numpy as np
import math 

x=math.pi/4
dxarr=[]
yarr=[]
for dx in 10**np.arange(-16, -0.5, 0.1):
    y=derifw(np.sin,x,dx)
    print('dx %g,  y %g'%(dx,y))
    dxarr.append(dx)
    yarr.append(y)
```

```python
import numpy as np
import matplotlib.pyplot as plt

plt.plot(dxarr, abs(yarr-np.cos(1.)), 'b-', lw=2)     
plt.xscale('log')
plt.yscale('log')
plt.xlabel('dx')
plt.ylabel('Error')
plt.show()
```
