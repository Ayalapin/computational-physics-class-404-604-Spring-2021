```python
# try this, calculate 1+2+3....+100

# EasyMatPlot.py:                 Simple use of matplotlib's plot command 
import pylab as p                           # import Matplotlib label it as p
from numpy import *                                 # import numpy library

p.title('Homework 1')              # Title and labels
p.xlabel('x - values')
p.ylabel('y - values')

delta = 0.1

x=p.arange( -pi, pi, delta )
y=1.

for i in range(1,12):
    y=y+2*(sin(i*x)/i) + pi/2
print(y)

plt.plot(x,y)

x0=p.arange( -pi, pi, delta )
y0=1.

for i in range(1,21):
    y0=y0+2*(sin(i*x)/i) + pi/2
print(y)

p.plot(x0, y0, 'bo')                          

x1=p.arange( -pi, pi, delta )
y1=1.

for i in range(1,31):
    y1=y1+2*(sin(i*x)/i) + pi/2
print(y)

p.plot(x1, y1, 'r')
plt.show()
```
