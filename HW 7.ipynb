```python
# LaplaceLine.py:  Solve Laplace's eqtn, 3D matplot, close shell to quit


import matplotlib.pylab as p;
import math 
from mpl_toolkits.mplot3d import Axes3D
from numpy import *;
import numpy;
print("Initializing")
omega=1
h=1
Nmax = 100; Niter = 5000; 
row = zeros((Nmax, Nmax), float)
V = zeros((Nmax, Nmax), float)  # potential at i and j, V(i,j), 100 x and 100 y, x0----x100, y0---y100 
V[1,:]=.0.0
V[:1]= 0.0
#V[:,1]= 0.0
row[50,50] = 1

for k in range(0, Nmax):  V[k,k] = -100            
    
for iter in range(Niter):                                  
    if iter%10 == 0: print(iter)
    maxerr=0.
    for i in range(1, Nmax-1):                                                
        for j in range(1,Nmax-1): 
            du = (V[i+1,j]+V[i-1,j]+V[i,j+1]+V[i,j-1])/4.-(h**2)*math.pi*row[i,j]-V[i,j] 
            if(maxerr<abs(du)):
                maxerr=abs(du) 
            V[i,j] = V[i,j] + omega*du    #du (error) it correct to new solution        # get the new V[i,j]
    print(maxerr)
    if(maxerr<1.e-3):
        break

x = range(0, Nmax, 1)
y = range(0, Nmax, 1)                              
X, Y = p.meshgrid(x,y)                 

def functz(V):          # V(x, y) 
    z = V[X,Y]-4*math.pi*row                        
    return z

Z = functz(V)                          
fig = p.figure()                                      # Create figure
ax = Axes3D(fig)                                      # Plot axes
ax.plot_wireframe(X, Y, Z, color = 'r')               # Red wireframe
ax.set_xlabel('X')                                     
ax.set_ylabel('Y')
ax.set_zlabel('Potential')
p.show()                                              # Show fig fig
```
